```python
def get_half_mass_redshift(id0, mass, id0_groups) :
    import numpy as np
    group = id0_groups[id0]
    zbins = np.arange(0, 10, 0.01)
    
    interp_m = np.interp(zbins,group['z'],group['m'])
    half_mass_index = np.where(interp_m < 0.5 * mass)
    try :
        return zbins[half_mass_index].min()   
    except ValueError :
        print "No half mass redshift stored"
        return zbins[np.where(group['m']==mass)].min()
```
